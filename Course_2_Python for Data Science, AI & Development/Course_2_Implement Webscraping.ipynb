```python
from bs4 import BeautifulSoup
import html5lib
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_banks?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkPY0221ENSkillsNetwork23455645-2022-01-01"

r = requests.get(url)
print("Status Code: ", r.status_code)
print("Headers: ", r.request.headers)
print("Body: ", r.request.body)
```

```python
header = r.headers
```

```python
print("Date: ", header['date'])
print("Content Type: ", header['Content-Type'])
print("Encoding: ", r.encoding)
```

```python
print(r.text[0:100])
```

```python
# Data in the web page
data = r.text
```

```python
data[101:124]
```

```python
# Pass the source code of the web page to beautiful soup to create a beautiful soup object for it.
soup = BeautifulSoup(data, 'html.parser')
```

```python
#print(soup.prettify())
```

```python
data = pd.DataFrame(columns=["Name", "Market Cap (US$ Billion)"])

for row in soup.find_all('tbody')[3].find_all('tr'):
    col = row.find_all('td')
    if len(col) > 0:
        name = col[1].text.strip()
        market_cap = float(col[2].get_text(strip=True)[:-3])
        data = data.append({"Name": name, "Market Cap (US$ Billion)": market_cap}, ignore_index=True)
```

```python
# load to json
data.to_json(r'./bank_market_cap.json', orient = 'index')
```
